```python
#Loading Country Code Dataset to map
df_countrycode = pd.read_excel("sample_data/Country-Code.xlsx")
df_countrycode
```

```python
df_countrycode.columns
```

```python
final_df = pd.merge(df,df_countrycode,on='Country Code' , how = "left" )
final_df.head(3)
```

```python
#find the order count
country_names =final_df.Country.value_counts()
country_values =final_df.Country.value_counts().index
country_values
country_names
```

```python
##Pie Chart
plt.pie(country_names[:3],labels = country_values[:3],autopct = "%1.2f%%")
```

##Observation
## Major Business is from India than other countries

```python
final_df.columns
```

Finding Missing Values

```python
#Counting the Rating Values
final_df.groupby(['Aggregate rating' ,'Rating color','Rating text']).size()
```

```python
#Renaming the column name
ratings =final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns= {0:'Rating Count'})
ratings
```

**Ratings Observation **
1)Ratings between 4.5 to 4.9 is Excellent
2)Rating between 4.0 to 4.4 is Very Good
3)Rating between 3.5 to 3.9 is Good

4)Rating between 2.5 to 3.4 is Average
5)Rating between 1.8 to 2.4 is poor
6)Customers not rated below 1.8

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,8)
sns.barplot( x = 'Aggregate rating' , y = 'Rating Count', data =ratings ,hue = 'Rating color', palette = ['Blue',"Red",'Orange','Yellow','Green','DarkGreen'])
```

Observation


1.   Not Rated count is very high
2.   Maximum no of rating is between 2.5 to 3.4





```python
##Count Value
sns.countplot(x = 'Rating color',data =ratings , palette = ['Blue',"Red",'Orange','Yellow','Green','DarkGreen'])
```
